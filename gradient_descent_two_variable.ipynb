```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining the function
def f(X):
    x, y = X
    return x**3 - 3*x*y**2 + y**2

#defining the derivatives
def grad_f(X):
    x, y = X
    return np.array([3*x**2 - 3*y**2, -6*x*y + 2*y])


#initiating the values
X = np.random.uniform(-1,1,2)
a = 0.001 #learning rate
iterations = 100

trajectory = [X.copy()]

#defining the gradient descent
for i in range(iterations):
    grad = grad_f(X)
    X = X - a*grad
    trajectory.append(X.copy())
    
trajectory = np.array(trajectory)
print(f'Minimum found at X = {X}, f(X) = {f(X)}')
```

```python
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X_grid, Y_grid = np.meshgrid(x_vals, y_vals)
Z = 2*X_grid**2*Y_grid**3

#3D Surface plot
fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X_grid, Y_grid, Z, cmap ='viridis', alpha = 0.8)
ax.plot(trajectory[:,0], trajectory[:,1], f(trajectory.T), color = 'r', marker ='o', label = 'Gradient Descent Path')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Gradient Descent on f(x,y) = 2x^2y^3')
plt.show()
```

```python
#2D contour plot
plt.figure(figsize = (7,6))
plt.contour(X_grid, Y_grid, Z, levels = 30, cmap = 'viridis')
plt.plot(trajectory[:,0], trajectory[:,1],'ro-', label = 'Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Path on Contour Plot')
plt.legend()
plt.show()
```
