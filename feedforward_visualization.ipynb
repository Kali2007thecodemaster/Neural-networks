```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#initializing the features
#input , 2 features, 3 neurons
np.random.seed(42)
W = np.random.rand(2,3)
b =np.random.rand(1,3)

print("Weights (W): \n", W)
print("Biases (b): \n", b)

#activation function
def sigmoid(z):
    return 1/ (1 +np.exp(-z))
```

```python
#Creating a grid of points
x1 = np.linspace(-2,2,100)
x2 = np.linspace(-2,2,100)
X1, X2 = np.meshgrid(x1, x2)
X_grid = np.c_[X1.ravel(), X2.ravel()]
```

```python
#FeedForward computation
Z = np.dot(X_grid, W) + b
A = sigmoid(Z)
```

```python
fig = plt.figure(figsize =(18,5))
for j in range (3):
    ax = fig.add_subplot(1, 3, j+1, projection = '3d')
    A_j = A[:, j].reshape(X1.shape)
    ax.plot_surface(X1, X2, A_j, cmap = 'viridis', alpha = 0.8)
    
    ax.set_title(f'Neuron {j+1}', fontsize = 12)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_zlabel('$a_{%d}$' % (j+1))
    ax.view_init(elev = 25, azim = 30)

plt.suptitle('3D Activation Surface of each neuron', fontsize = 15)
plt.tight_layout()
plt.show()
```
