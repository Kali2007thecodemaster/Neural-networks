```python
#importing importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 3*(x-x**2)**3 + 7

def df(x):
    return 9*((x**2-x))*(2*x-1)

# Initializing the values
x = np.random.rand() * 10  # Random initial point
a = 0.1 #learning rate
iterations = 50

# Gradient Descent Algorithm
for i in range(iterations):
    grad = df(x)
    x = x - a*grad
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")
    
print(f"the minimum found at x = {x}, f(x) = {f(x)}")
```

```python
#plotting the gradient descent
x_vals = np.linspace(-1, 5, 100)
plt.plot(x_vals, f(x_vals))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('gradient descent')
plt.axvline(x=x, color='r', linestyle = '--', label = f'Minimum at ({x:.2f},{f(x):.2f}) ')
plt.legend()
plt.show()
```
