This is a simple implementation of a  perceptron using sigmoid activation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining an array of inputs x
x = np.random.rand(3, 3)
#defining an array of weights w
w = np.random.rand(3, 1)
#defining a bias b
b = np.random.rand(3, 1)

#defining the sigmoid activation function
def sigmoid(z):
    return 1/(1 + np.exp(-z))

#calculating the weighted sum z
z = np.dot(x, w) + b

#calculating the output of the neuron y1 using the sigmoid activation function
y1 = sigmoid(z)

#calculating the output of the neuron y2 using softmax activation function
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # for numerical stability
    return exp_z / exp_z.sum(axis=0)

#calculating the output of the neuron y2 using softmax activation function
y2 = softmax(z)

#printing the outputs
print("Output using Sigmoid activation:", y1)
print("Output using Softmax activation:", y2)

plt.plot(y1, label='Sigmoid Output')
plt.plot(y2, label='Softmax Output')
plt.legend()
plt.show()
```
